```python
# Robust auto-discovery loader for PhonePe Pulse JSON -> SQLite
from pathlib import Path
import json, sqlite3, itertools

ROOT = Path("data")
print("Root data folder:", ROOT.resolve())
all_jsons = list(ROOT.rglob("*.json"))
print("Total .json files found under data/:", len(all_jsons))

# helper: try several possible places inside JSON where transaction records could live
def extract_transaction_list(obj):
    """
    Return a list of transaction-record dicts, or None if can't find.
    Common PhonePe shapes:
      obj["data"]["transactionData"]
      obj["data"]["transactions"]
      obj.get("data") being list
      obj["transactionData"]
    """
    if not isinstance(obj, dict):
        return None
    # common variants
    if "data" in obj:
        d = obj["data"]
        if isinstance(d, dict):
            for key in ("transactionData", "transactions", "transaction"):
                if key in d and isinstance(d[key], (list,tuple)):
                    return d[key]
        # sometimes d is a list of sections
        if isinstance(d, list):
            # find first element that contains transactionData or paymentInstruments
            for el in d:
                if isinstance(el, dict):
                    if "transactionData" in el and isinstance(el["transactionData"], list):
                        return el["transactionData"]
                    # sometimes nested: el["data"]["transactionData"]
                    if "data" in el and isinstance(el["data"], dict):
                        if "transactionData" in el["data"]:
                            return el["data"]["transactionData"]
    # fallback top-level keys
    for key in ("transactionData","transactions","paymentData"):
        if key in obj and isinstance(obj[key], (list,tuple)):
            return obj[key]
    return None

# helper: derive state name from file path by searching up for a folder that looks like a state
def infer_state_from_path(p: Path):
    # look for a parent named exactly "state" then take next child name before file
    for parent in p.parents:
        if parent.name.lower() == "state":
            # the folder just below 'state' in the path is the state folder (i.e. parent_of_file -> .../state/<state_name>/.../<file>)
            try:
                # find index of this parent in parents and take the previous one in the chain
                parts = list(p.resolve().parts)
                # find last occurrence of 'state' (case-insensitive)
                idxs = [i for i,s in enumerate(parts) if s.lower() == "state"]
                if idxs:
                    idx = idxs[-1]
                    # the state folder name is parts[idx+1] if exists
                    if idx+1 < len(parts):
                        return parts[idx+1].lower()
            except Exception:
                pass
    # fallback: use the immediate parent folder's name (may work if files are stored under state/<state>/<file>)
    return p.parent.name.lower()

# prepare DB
DB = Path("phonepe.db")
conn = sqlite3.connect(DB)
cur = conn.cursor()
cur.execute("""
CREATE TABLE IF NOT EXISTS aggregated_transaction (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    year INTEGER,
    quarter INTEGER,
    state TEXT,
    transaction_type TEXT,
    transaction_count INTEGER,
    transaction_amount REAL,
    source_file TEXT
);
""")
conn.commit()

inserted = 0
failed_files = []

# iterate all json files
for i, jf in enumerate(all_jsons, start=1):
    try:
        with open(jf, "r", encoding="utf-8") as f:
            obj = json.load(f)
    except Exception as e:
        failed_files.append((str(jf), f"json-load-failed: {e}"))
        continue

    tx_list = extract_transaction_list(obj)
    if not tx_list:
        # not transaction file; skip
        continue

    # infer year/quarter from filename (best-effort)
    fname = jf.name
    year = None
    quarter = None
    # look for pattern like 2018-Q1 or 2018_Q1 or 2018-01
    import re
    m = re.search(r"(20\d{2})[-_ ]?Q?([1-4])", fname, re.IGNORECASE)
    if m:
        year = int(m.group(1))
        quarter = int(m.group(2))
    else:
        # try year only
        m2 = re.search(r"(20\d{2})", fname)
        if m2:
            year = int(m2.group(1))

    state = infer_state_from_path(jf)

    # Some transaction items have structure with "name" and "paymentInstruments"
    # We'll try to read common fields safely
    wrote_any = False
    for item in tx_list:
        if not isinstance(item, dict):
            continue
        # find 'name'
        name = item.get("name") or item.get("title") or item.get("key") or None

        # find paymentInstruments list inside item
        pis = None
        if "paymentInstruments" in item and isinstance(item["paymentInstruments"], list):
            pis = item["paymentInstruments"]
        else:
            # sometimes structure: item -> "data" -> "paymentInstruments"
            if "data" in item and isinstance(item["data"], dict) and "paymentInstruments" in item["data"]:
                pis = item["data"]["paymentInstruments"]

        # if pis is a list, attempt to take first element metrics
        count = None
        amount = None
        if isinstance(pis, list) and pis:
            p0 = pis[0]
            if isinstance(p0, dict):
                count = p0.get("count") or p0.get("totalCount") or p0.get("transactions") or None
                amount = p0.get("amount") or p0.get("totalAmount") or p0.get("amountIn") or None

        # Some files may put metrics directly under item, e.g., item.get('count') etc.
        if count is None:
            for k in ("count","txnCount","transactions"):
                if k in item:
                    count = item.get(k)
                    break
        if amount is None:
            for k in ("amount","txnAmount","totalAmount"):
                if k in item:
                    amount = item.get(k)
                    break

        # coercions
        try:
            if count is not None:
                count = int(count)
        except Exception:
            try:
                count = int(float(count))
            except:
                count = None
        try:
            if amount is not None:
                amount = float(amount)
        except Exception:
            amount = None

        # If we have at least a name and one metric, insert
        if name is not None and (count is not None or amount is not None):
            cur.execute("""
                INSERT INTO aggregated_transaction (year, quarter, state, transaction_type, transaction_count, transaction_amount, source_file)
                VALUES (?, ?, ?, ?, ?, ?, ?)
            """, (year, quarter, state, name, count, amount, str(jf)))
            inserted += 1
            wrote_any = True

    if not wrote_any:
        # we found transaction list but couldn't extract entries -> log for review
        failed_files.append((str(jf), "transaction-list-found-but-no-extractable-metrics"))

    # commit periodically
    if i % 500 == 0:
        conn.commit()

# final commit & close
conn.commit()
conn.close()

print("Done. JSON files scanned:", len(all_jsons))
print("Rows inserted:", inserted)
print("Files with issues (sample 10):", len(failed_files))
for f in failed_files[:10]:
    print("-", f)

```

```python
# Resume-safe loader: will skip files already recorded in DB
from pathlib import Path
import json, sqlite3, re
from tqdm.notebook import tqdm

ROOT = Path(r"C:\Users\USER\Desktop\phonepay\pulse")  # your path
all_jsons = list(ROOT.rglob("*.json"))
print("Total JSON files found:", len(all_jsons))

# Helper functions (same as before)
def extract_transaction_list(obj):
    if not isinstance(obj, dict): return None
    if "data" in obj:
        d = obj["data"]
        if isinstance(d, dict):
            for key in ("transactionData","transactions","transaction"):
                if key in d and isinstance(d[key], (list,tuple)):
                    return d[key]
        if isinstance(d, list):
            for el in d:
                if isinstance(el, dict):
                    if "transactionData" in el and isinstance(el["transactionData"], list):
                        return el["transactionData"]
                    if "paymentInstruments" in el and isinstance(el.get("paymentInstruments"), list):
                        return d
    for k in ("transactionData","transactions","paymentData"):
        if k in obj and isinstance(obj[k], (list,tuple)):
            return obj[k]
    return None

def infer_state_from_path(p: Path):
    parts = [s.lower() for s in p.resolve().parts]
    if "state" in parts:
        idxs = [i for i,s in enumerate(parts) if s=="state"]
        if idxs:
            idx = idxs[-1]
            if idx+1 < len(parts):
                return parts[idx+1]
    return p.parent.name.lower()

def infer_year_quarter_from_filename(fname: str):
    year = None; quarter = None
    m = re.search(r"(20\d{2})[-_ ]?Q?([1-4])", fname, re.IGNORECASE)
    if m:
        year = int(m.group(1)); quarter = int(m.group(2))
    else:
        m2 = re.search(r"(20\d{2})", fname)
        if m2:
            year = int(m2.group(1))
    return year, quarter

# DB connection
DB = Path("phonepe.db")
conn = sqlite3.connect(DB)
cur = conn.cursor()

# Load set of already processed files (source_file)
cur.execute("SELECT DISTINCT source_file FROM aggregated_transaction WHERE source_file IS NOT NULL;")
existing = set(r[0] for r in cur.fetchall())
print("Already recorded files in DB:", len(existing))

inserted = 0
skipped_files = []

for jf in tqdm(all_jsons, desc="Scanning JSONs"):
    jf_str = str(jf)
    # skip files already processed
    if jf_str in existing:
        continue

    try:
        with open(jf, "r", encoding="utf-8") as f:
            obj = json.load(f)
    except Exception as e:
        skipped_files.append((jf_str, f"json-load-failed: {e}"))
        continue

    tx_list = extract_transaction_list(obj)
    if not tx_list:
        # not a transaction file (or structure not recognized)
        continue

    year, quarter = infer_year_quarter_from_filename(jf.name)
    state = infer_state_from_path(jf)
    wrote_any = False

    for item in tx_list:
        if not isinstance(item, dict):
            continue
        name = item.get("name") or item.get("title") or item.get("key") or item.get("category") or None

        pis = None
        if "paymentInstruments" in item and isinstance(item["paymentInstruments"], list):
            pis = item["paymentInstruments"]
        elif "data" in item and isinstance(item["data"], dict) and "paymentInstruments" in item["data"]:
            pis = item["data"]["paymentInstruments"]

        count = None; amount = None
        if isinstance(pis, list) and pis:
            p0 = pis[0]
            if isinstance(p0, dict):
                count = p0.get("count") or p0.get("transactions") or p0.get("totalCount")
                amount = p0.get("amount") or p0.get("totalAmount")
        if count is None:
            for k in ("count","txnCount","transactions","totalCount"):
                if k in item:
                    count = item.get(k); break
        if amount is None:
            for k in ("amount","txnAmount","totalAmount"):
                if k in item:
                    amount = item.get(k); break

        # coerce
        try:
            if count is not None: count = int(count)
        except: 
            try: count = int(float(count))
            except: count = None
        try:
            if amount is not None: amount = float(amount)
        except:
            amount = None

        if name is not None and (count is not None or amount is not None):
            cur.execute("""
                INSERT INTO aggregated_transaction
                (year, quarter, state, transaction_type, transaction_count, transaction_amount, source_file)
                VALUES (?, ?, ?, ?, ?, ?, ?)
            """, (year, quarter, state, name, count, amount, jf_str))
            inserted += 1
            wrote_any = True

    if not wrote_any:
        skipped_files.append((jf_str, "transaction-list-present-but-no-extractable-metrics"))

    # commit periodically
    if inserted and inserted % 1000 == 0:
        conn.commit()
        # update existing set to avoid reprocessing if restarted
        cur.execute("SELECT DISTINCT source_file FROM aggregated_transaction WHERE source_file IS NOT NULL;")
        existing = set(r[0] for r in cur.fetchall())

# final commit & close
conn.commit()
conn.close()

print("Completed. New rows inserted in this run:", inserted)
print("Skipped/problem files count (examples up to 10):", len(skipped_files))
for s in skipped_files[:10]:
    print("-", s)
```

```python
# Backfill year/quarter from source_file paths in DB
import sqlite3
from pathlib import Path
import re

DB = "phonepe.db"
conn = sqlite3.connect(DB)
cur = conn.cursor()

# Get rows needing backfill
cur.execute("SELECT id, source_file, year, quarter FROM aggregated_transaction WHERE (year IS NULL OR quarter IS NULL) AND source_file IS NOT NULL")
rows = cur.fetchall()
print("Rows to inspect for backfill:", len(rows))

fixed = 0
left = 0
examples_not_fixed = []

def try_infer(path_str):
    p = Path(path_str)
    parts = [part for part in p.parts if part]  # full parts
    # Normalize parts to plain strings (lower for comparisons)
    parts_low = [str(x).lower() for x in parts]

    # Year: first 4-digit 20xx in parts (search from left to right)
    year = None
    for part in parts:
        m = re.search(r"(20\d{2})", str(part))
        if m:
            year = int(m.group(1))
            break

    # Quarter: search for Q1..Q4 or folder named '1'..'4'
    quarter = None
    # Check for explicit q like 'Q1' or 'q1'
    for part in parts_low:
        mq = re.match(r"q([1-4])$", part)
        if mq:
            quarter = int(mq.group(1))
            break
    if quarter is None:
        # look for numeric folder names '1','2','3','4' â€” often parent folder is quarter
        for part in parts:
            if str(part).isdigit() and int(part) in (1,2,3,4):
                quarter = int(part)
                break
    # fallback: if filename itself is '1.json' etc
    fname = parts[-1] if parts else ""
    if quarter is None:
        mfn = re.match(r"([1-4])(\.json)?$", str(fname))
        if mfn:
            quarter = int(mfn.group(1))

    return year, quarter

# We'll update rows in batches to avoid long transaction locks
to_update = []
for row in rows:
    rid, source_file, year, quarter = row
    y, q = try_infer(source_file)
    # If we didn't find year/quarter, leave as None
    # Only update columns that are currently NULL to avoid overwriting valid values
    new_year = year if (year is None or year==None) and y is not None else (year if year is not None else y)
    new_quarter = quarter if (quarter is None or quarter==None) and q is not None else (quarter if quarter is not None else q)
    # But simpler: if DB has null and we inferred non-null, set it
    set_year = None
    set_quarter = None
    if year is None and y is not None:
        set_year = y
    if quarter is None and q is not None:
        set_quarter = q
    if set_year is not None or set_quarter is not None:
        to_update.append((set_year, set_quarter, rid))
    else:
        examples_not_fixed.append((rid, source_file))

# Apply updates
for set_year, set_quarter, rid in to_update:
    if set_year is not None and set_quarter is not None:
        cur.execute("UPDATE aggregated_transaction SET year = ?, quarter = ? WHERE id = ?", (set_year, set_quarter, rid))
    elif set_year is not None:
        cur.execute("UPDATE aggregated_transaction SET year = ? WHERE id = ?", (set_year, rid))
    elif set_quarter is not None:
        cur.execute("UPDATE aggregated_transaction SET quarter = ? WHERE id = ?", (set_quarter, rid))
    fixed += 1

conn.commit()

# report
print("Rows updated (attempted):", len(to_update))
print("Examples not fixed (first 10):", len(examples_not_fixed))
for ex in examples_not_fixed[:10]:
    print("-", ex)

# How many still missing?
cur.execute("SELECT COUNT(*) FROM aggregated_transaction WHERE year IS NULL OR quarter IS NULL")
remain = cur.fetchone()[0]
print("Rows still missing year/quarter after backfill:", remain)

conn.close()
```

```python
import streamlit as st
import sqlite3
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="PhonePe Transaction Insights", layout="wide")

conn = sqlite3.connect("phonepe.db")

# Top states
df_states = pd.read_sql_query("""
SELECT state, SUM(transaction_amount) AS total_amount
FROM aggregated_transaction
WHERE state NOT GLOB '[0-9]*'
GROUP BY state
ORDER BY total_amount DESC
LIMIT 15;
""", conn)

# Trend
trend = pd.read_sql_query("""
SELECT year, quarter, SUM(transaction_amount) AS total_amount
FROM aggregated_transaction
WHERE year IS NOT NULL AND quarter IS NOT NULL
GROUP BY year, quarter
ORDER BY year, quarter;
""", conn)

# Transaction types
df_types = pd.read_sql_query("""
SELECT transaction_type, SUM(transaction_amount) AS total_amount
FROM aggregated_transaction
WHERE transaction_type NOT NULL
GROUP BY transaction_type
ORDER BY total_amount DESC;
""", conn)

conn.close()

st.title("ðŸ“Š PhonePe Pulse Transaction Insights Dashboard")

st.header("Top States by Transaction Amount")
fig1 = px.bar(df_states, x="state", y="total_amount", color="total_amount")
st.plotly_chart(fig1, use_container_width=True)

st.header("Quarterly Transaction Trend")
fig2 = px.line(trend, x=trend.index, y="total_amount", markers=True)
st.plotly_chart(fig2, use_container_width=True)

st.header("Transaction Categories")
fig3 = px.pie(df_types, names="transaction_type", values="total_amount")
st.plotly_chart(fig3, use_container_width=True)
```
